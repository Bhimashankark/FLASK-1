Q1. What is Flask Framework? What are the advantages of Flask Framework?

ANSWER : Flask is a popular Python-based micro web framework that allows developers to quickly build web applications.

Some advantages of using Flask framework are:

Lightweight and flexible: Flask is a lightweight framework that allows developers to customize and tailor their web applications as per their requirements. It provides only the basic tools and libraries required to build web applications, which makes it very flexible.

Easy to use: Flask has a simple and intuitive API that is easy to understand and use, even for beginners. It provides a clear and concise structure for organizing code, making it easy to maintain and update.

Extensible: Flask provides a wide range of extensions and libraries that can be used to add extra functionality to the application, such as authentication, database connectivity, and testing.

Integrated with the Werkzeug toolkit and Jinja2 templating engine: Flask is built on top of the Werkzeug toolkit, which provides robust HTTP request and response handling. It also uses the Jinja2 templating engine, which allows developers to create dynamic and interactive web pages.

Supports various data storage solutions: Flask supports various data storage solutions, including SQL databases, NoSQL databases, and file-based databases.

Ideal for small to medium-sized applications: Flask is an excellent choice for small to medium-sized web applications that do not require complex features or scalability.

Q3. What is App routing in Flask? Why do we use app routes?

ANSWER : In Flask, app routing refers to the process of defining URL patterns (or routes) that correspond to different parts of the web application. These routes are used to map specific URLs to their respective functions or views that handle the HTTP requests and generate HTTP responses.

App routing is an essential feature of Flask because it helps to organize the application's code and makes it easier to navigate. By defining routes, developers can ensure that users can access different parts of the application using specific URLs. This can also help in implementing a clean and user-friendly interface, where URLs are easy to remember and navigate.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

ANSWER : In Flask, the url_for() function is used for URL building. This function takes the name of the view function as its first argument and generates a URL for that function based on the routing information provided in the Flask application. It can also take additional arguments that will be used as variables in the URL.

Here is an example code that demonstrates how to use the url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/user/<username>')
def user_profile(username):
    return 'Profile page of %s' % username

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return 'Post %d' % post_id

with app.test_request_context():
    print(url_for('home'))
    print(url_for('user_profile', username='John Doe'))
    print(url_for('show_post', post_id=1))
```
